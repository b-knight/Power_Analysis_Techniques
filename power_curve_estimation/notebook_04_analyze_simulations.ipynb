```python
df1    = pd.read_csv("df1.csv") 
results_1 = df1['Actual Power'].tolist()
df2    = pd.read_csv("df2.csv") 
results_2 = df2['Actual Power'].tolist()
df3    = pd.read_csv("df3.csv") 
results_3 = df3['Actual Power'].tolist()
df4    = pd.read_csv("df4.csv") 
results_4 = df4['Actual Power'].tolist()
df5    = pd.read_csv("df5.csv") 
results_5 = df5['Actual Power'].tolist()
df6    = pd.read_csv("df6.csv") 
results_6 = df6['Actual Power'].tolist()
df7    = pd.read_csv("df7.csv") 
results_7 = df7['Actual Power'].tolist()
df8    = pd.read_csv("df8.csv") 
results_8 = df8['Actual Power'].tolist()
df9    = pd.read_csv("df9.csv") 
results_9 = df9['Actual Power'].tolist()

data_to_plot = [results_1,results_2,results_3,results_4,results_5,results_6,results_7,results_8,results_9]
```

```python
import matplotlib as mpl 
mpl.use('agg')
%matplotlib inline
import matplotlib.pyplot as plt 
from matplotlib import rc, rcParams
rcParams['text.latex.preamble'] = [r'\usepackage{sfmath} \boldmath']
rc('text', usetex=True)
rc('axes', linewidth=2)
rc('font', weight='bold')
rcParams['text.latex.preamble'] = [r'\usepackage{sfmath} \boldmath']
## add patch_artist=True option to ax.boxplot() 
## to get fill color

# Create a figure instance
fig = plt.figure(1, figsize=(16, 6))

# Create an axes instance
ax = fig.add_subplot(111)

bp = ax.boxplot(data_to_plot, patch_artist=True)

## change outline color, fill color and linewidth of the boxes
for box in bp['boxes']:
    # change outline color
    box.set( color='#7570b3', linewidth=2)
    # change fill color
    box.set( facecolor = '#1b9e77' )

## change color and linewidth of the whiskers
for whisker in bp['whiskers']:
    whisker.set(color='#7570b3', linewidth=2)

## change color and linewidth of the caps
for cap in bp['caps']:
    cap.set(color='#7570b3', linewidth=2)

## change color and linewidth of the medians
for median in bp['medians']:
    median.set(color='#b2df8a', linewidth=2)

## change the style of fliers and their fill
for flier in bp['fliers']:
    flier.set(marker='o', color='#e7298a', alpha=0.5)
ax.axhline(y=0.8, xmin=0.0, xmax=1.0, color='r')    

## Custom x-axis labels
ax.tick_params(axis='y', which='major', labelsize=16)
ax.tick_params(axis='x', which='major', labelsize=16)
label_1 = r"\textbf{Method I}" "\n" "$\mu \pm[1]\sigma$" "\n" "100 Iterations" "\n" " per Point" \
           "\n" "X 3" "\n" "Total" "\n" "Iterations" "\n" "= 800"
label_2 = r"\textbf{Method II}" "\n" "$\mu \pm[1,2]\sigma$" "\n" "100 Iterations" "\n" " per Point" \
           "\n" "X 5" "\n" "Total" "\n" "Iterations" "\n" "= 1,000"
label_3 = r"\textbf{Method III}" "\n" "$\mu \pm[1,2,3]\sigma$" "\n" "100 Iterations" "\n" " per Point" \
           "\n" "X 7" "\n" "Total" "\n" "Iterations" "\n" "= 1,200"
label_4 = r"\textbf{Method IV}" "\n" "$\mu \pm[1]\sigma$" "\n" "200 Iterations" "\n" " per Point" \
           "\n" "X 3" "\n" "Total" "\n" "Iterations" "\n" "= 1,100"
label_5 = r"\textbf{Method V}" "\n" "$\mu \pm[1,2]\sigma$" "\n" "200 Iterations" "\n" " per Point" \
           "\n" "X 5" "\n" "Total" "\n" "Iterations" "\n" "= 1,500"
label_6 = r"\textbf{Method VI}" "\n" "$\mu \pm[1,2,3]\sigma$" "\n" "200 Iterations" "\n" " per Point" \
           "\n" "X 7" "\n" "Total" "\n" "Iterations" "\n" "= 1,900"
label_7 = r"\textbf{Method VII}" "\n" "$\mu \pm[1]\sigma$" "\n" "300 Iterations" "\n" " per Point" \
           "\n" "X 3" "\n" "Total" "\n" "Iterations" "\n" "= 1,400"
label_8 = r"\textbf{Method VIII}" "\n" "$\mu \pm[1,2]\sigma$" "\n" "300 Iterations" "\n" " per Point" \
           "\n" "X 5" "\n" "Total" "\n" "Iterations" "\n" "= 2,000"
label_9 = r"\textbf{Method IX}" "\n" "$\mu \pm[1,2,3]\sigma$" "\n" "300 Iterations" "\n" " per Point" \
           "\n" "X 7" "\n" "Total" "\n" "Iterations" "\n" "= 2,600"
ax.set_xticklabels([label_1,label_2,label_3,label_4,label_5,label_6,label_7,label_8,label_9])

# Y=Axis
plt.ylabel(r"\textbf{Estimated} \\\\ \phantom{aaaa}\textbf{Power:} \\\\ \phantom{aa}$\%$ \textbf{of 500} \\\\ "
           + r"\phantom{a}\textbf{Simulations} \\\\ \textbf{where NULL} \\\\ \textbf{was Rejected} \\\\" +
           r"\phantom{aa}$\alpha < 0.05$",
           rotation = 0, 
           labelpad = 70,
           size = 16)

ax.yaxis.set_label_coords(-0.10,0.25)
# Plot Title
plt.title("Distributions of Power Estimates Attained Through Inference of the Power Curve", size=28)

# Save the figure
fig.savefig('nine_model_results.png', bbox_inches='tight')
```
