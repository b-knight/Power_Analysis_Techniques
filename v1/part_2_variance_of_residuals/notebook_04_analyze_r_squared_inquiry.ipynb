```python
import pandas as pd
```

```python
df1a = pd.read_csv("./residual_dfs/results/mde_0_05_n_100000_model_I.csv") 
df1b = pd.read_csv("./residual_dfs/results/mde_0_05_n_100000_model_II.csv")
df1c = pd.read_csv("./residual_dfs/results/mde_0_05_n_100000_model_III.csv") 
df1d = pd.read_csv("./residual_dfs/results/mde_0_05_n_100000_model_IV.csv")
df2a = pd.read_csv("./residual_dfs/results/mde_0_025_n_100000_model_I.csv") 
df2b = pd.read_csv("./residual_dfs/results/mde_0_025_n_100000_model_II.csv")
df2c = pd.read_csv("./residual_dfs/results/mde_0_025_n_100000_model_III.csv") 
df2d = pd.read_csv("./residual_dfs/results/mde_0_025_n_100000_model_IV.csv")
df3a = pd.read_csv("./residual_dfs/results/mde_0_01_n_100000_model_I.csv") 
df3b = pd.read_csv("./residual_dfs/results/mde_0_01_n_100000_model_II.csv")
df3c = pd.read_csv("./residual_dfs/results/mde_0_01_n_100000_model_III.csv") 
df3d = pd.read_csv("./residual_dfs/results/mde_0_01_n_100000_model_IV.csv")
df4a = pd.read_csv("./residual_dfs/results/mde_0_005_n_100000_model_I.csv") 
df4b = pd.read_csv("./residual_dfs/results/mde_0_005_n_100000_model_II.csv")
df4c = pd.read_csv("./residual_dfs/results/mde_0_005_n_100000_model_III.csv") 
df4d = pd.read_csv("./residual_dfs/results/mde_0_005_n_100000_model_IV.csv")
df = pd.concat([df1a, df1b], axis=0)
df = pd.concat([df,   df1c], axis=0)
df = pd.concat([df,   df1d], axis=0)
df = pd.concat([df,   df2a], axis=0)
df = pd.concat([df,   df2b], axis=0)
df = pd.concat([df,   df2c], axis=0)
df = pd.concat([df,   df2d], axis=0)
df = pd.concat([df,   df3a], axis=0)
df = pd.concat([df,   df3b], axis=0)
df = pd.concat([df,   df3c], axis=0)
df = pd.concat([df,   df3d], axis=0)
df = pd.concat([df,   df4a], axis=0)
df = pd.concat([df,   df4b], axis=0)
df = pd.concat([df,   df4c], axis=0)
df = pd.concat([df,   df4d], axis=0)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.nonparametric.smoothers_lowess import lowess
import matplotlib.patches as patches

import matplotlib as mpl 
mpl.use('agg')
%matplotlib inline

import matplotlib.pyplot as plt 
from matplotlib import rc, rcParams

rcParams['text.latex.preamble'] = [r'\usepackage{sfmath} \boldmath']
rc('text', usetex=True)
rc('axes', linewidth=2)
rc('font', weight='bold')

fig = plt.figure(1, figsize=(12, 5))
ax  = fig.add_subplot(111)


x = df['Mean_R_Sqr']
y = df['Delta']

area = np.pi*25
plt.scatter(x, y, s=area, alpha=0.05)
plt.xlabel(r"\textbf{Adjusted R-Squared}", size = 18)
plt.ylabel(r" \phantom{aa}\textbf{Delta:} \\\\ \phantom{aa}Desired \\\\ \phantom{aa}Power - \\\\"
           + r" \phantom{a}Achieved \\\\ \phantom{aa}Power",
           rotation = 0, 
           labelpad = 60,
           size = 20)

ax.yaxis.set_label_coords(-0.12,0.4)
ax.tick_params(axis='y', which='major', labelsize=16)
ax.tick_params(axis='x', which='major', labelsize=16)

# smoothed = lowess(y, x, is_sorted=False, delta=0.01, frac=1.0, it=0)
# plt.plot(smoothed[:,0], smoothed[:,1], 'black', ls='--')

plt.figtext(.5,0.98,'Impact of Multicollinearity on Power Estimation', 
            fontsize=30, ha='center')
plt.figtext(.5,.91,'Inappropriate model specification leads to drastic underestimation of required sample size.',
            fontsize=20, ha='center')

rect1 = patches.Rectangle((0.92,-0.78),0.1,0.5, linewidth=1, edgecolor='r',facecolor='none')
rect2 = patches.Rectangle((-0.025, -0.19),0.05,0.42, linewidth=1, edgecolor='r',facecolor='none')
rect3 = patches.Rectangle((0.35, 0.06),0.025, 0.17, linewidth=1, edgecolor='r',facecolor='none')
rect4 = patches.Rectangle((0.38, 0.10),0.13, 0.13, linewidth=1, edgecolor='r',facecolor='none')
ax.add_patch(rect1)
ax.add_patch(rect2)
ax.add_patch(rect3)
ax.add_patch(rect4)


plt.annotate(r"\textbf{Model IV}",
             xy=(160, 190), 
             xycoords='figure pixels',
             fontsize=14)

plt.annotate(r"\textbf{Model III}",
             xy=(345, 255), 
             xycoords='figure pixels',
             fontsize=14)

plt.annotate(r"\textbf{Model II}",
             xy=(400, 268), 
             xycoords='figure pixels',
             fontsize=14)

plt.annotate(r"\textbf{Model I}",
             xy=(720, 190), 
             xycoords='figure pixels',
             fontsize=14)

# plt.show()
fig.savefig('multicollinearity_impact.png', bbox_inches='tight')
```
