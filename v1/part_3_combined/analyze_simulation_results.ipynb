```python
import matplotlib as mpl 
mpl.use('agg')
%matplotlib inline
import matplotlib.pyplot as plt 
from matplotlib import rc, rcParams
rcParams['text.latex.preamble'] = [r'\usepackage{sfmath} \boldmath']
rc('text', usetex=True)
rc('axes', linewidth=2)
rc('font', weight='bold')
rcParams['text.latex.preamble'] = [r'\usepackage{sfmath} \boldmath']
## add patch_artist=True option to ax.boxplot() 
## to get fill color

# Create a figure instance
fig = plt.figure(1, figsize=(16, 6))

# Create an axes instance
ax = fig.add_subplot(111)

bp = ax.boxplot(data_to_plot, patch_artist=True)

## change outline color, fill color and linewidth of the boxes
for box in bp['boxes']:
    # change outline color
    box.set( color='#7570b3', linewidth=2)
    # change fill color
    box.set( facecolor = '#00b3b3' )

## change color and linewidth of the whiskers
for whisker in bp['whiskers']:
    whisker.set(color='#7570b3', linewidth=2)

## change color and linewidth of the caps
for cap in bp['caps']:
    cap.set(color='#7570b3', linewidth=2)

## change color and linewidth of the medians
for median in bp['medians']:
    median.set(color='#b2df8a', linewidth=2)

## change the style of fliers and their fill
for flier in bp['fliers']:
    flier.set(marker='o', color='#e7298a', alpha=0.5)
ax.axhline(y=0.8, xmin=0.0, xmax=1.0, color='r')    

## Custom x-axis labels
ax.tick_params(axis='y', which='major', labelsize=16)
ax.tick_params(axis='x', which='major', labelsize=16)


label_1  = r"\textbf{Method}" "\n" "V"
label_2  = r"\textbf{Method}" "\n" "VI"
label_3  = r"\textbf{Method}" "\n" "VII"
label_4  = r"\textbf{Method}" "\n" "VIII"
label_5  = r"\textbf{Method}" "\n" "IX"
label_6  = r"\textbf{Method}" "\n" "X"
label_7  = r"\textbf{Method}" "\n" "XI"
label_8  = r"\textbf{Method}" "\n" "XII"
label_9  = r"\textbf{Method}" "\n" "XIII"
label_10 = r"\textbf{Method}" "\n" "XIV"
label_11 = r"\textbf{Method}" "\n" "XV"
label_12 = r"\textbf{Method}" "\n" "XVI"

ax.set_xticklabels([label_1,label_2,label_3,label_4,label_5,label_6,
                    label_7,label_8,label_9,label_10,label_11,label_12])

# Y=Axis
plt.ylabel(r"\textbf{Estimated} \\\\ \phantom{aaaa}\textbf{Power:} \\\\ \phantom{aa}$\%$ \textbf{of 500} \\\\ "
           + r"\phantom{a}\textbf{Simulations} \\\\ \textbf{where NULL} \\\\ \textbf{was Rejected} \\\\" +
           r"\phantom{aa}$\alpha < 0.05$",
           rotation = 0, 
           labelpad = 70,
           size = 16)

ax.yaxis.set_label_coords(-0.10,0.25)
# Plot Title
# plt.title(r"""\Huge{Big title !} \newline \normalsize{Small subtitle !}""")
plt.figtext(.5,0.95,'Distributions of Estimated Statistical Power (Combined Methods)', 
            fontsize=36, ha='center')
plt.figtext(.5,.9,'Each distribution represents 100 iterations.' +\
            ' A single iteration uses 900 simulations. ',
            fontsize=20, ha='center')


ax.axvline(x=4.5, color='black', ls='--', lw=1)  
ax.axvline(x=8.5, color='black', ls='--', lw=1)  

############################################################################################################
# Annotations of Models
############################################################################################################
plt.annotate(r"\textbf{Model II:}",
             xy=(191, 155), 
             xycoords='figure pixels',
             fontsize=14)
plt.annotate("Order Amount =",
             xy=(191, 135), 
             xycoords='figure pixels',
             fontsize=14)
plt.annotate("Treated +",
             xy=(191, 115), 
             xycoords='figure pixels',
             fontsize=14)
plt.annotate("Retailer Mean Order Amount +",
             xy=(191, 95), 
             xycoords='figure pixels',
             fontsize=14)
plt.annotate("Mean DOW Amount",
             xy=(191, 75), 
             xycoords='figure pixels',
             fontsize=14)
############################################################################################################
plt.annotate(r"\textbf{Model III:}",
             xy=(605, 280), 
             xycoords='figure pixels',
             fontsize=14)
plt.annotate("Order Amount =",
             xy=(605, 260), 
             xycoords='figure pixels',
             fontsize=14)
plt.annotate("Treated +",
             xy=(605, 240), 
             xycoords='figure pixels',
             fontsize=14)
plt.annotate("Mean DOW Amount",
             xy=(605, 220), 
             xycoords='figure pixels',
             fontsize=14)
############################################################################################################
plt.annotate(r"\textbf{Model IV:}",
             xy=(830, 280), 
             xycoords='figure pixels',
             fontsize=14)
plt.annotate("Order Amount =",
             xy=(830, 260), 
             xycoords='figure pixels',
             fontsize=14)
plt.annotate("Treated",
             xy=(830, 240), 
             xycoords='figure pixels',
             fontsize=14)
############################################################################################################
# Annotations of MDE
############################################################################################################
plt.annotate(r"\textbf{Absolute Treatment}",
             xy=(0, 20), 
             xycoords='figure pixels',
             fontsize=16)
plt.annotate(r"\textbf{Effect (MDE) =}",
             xy=(15, 5), 
             xycoords='figure pixels',
             fontsize=16)


plt.annotate(r"\textbf{\$4.67}",
             xy=(167, 5), 
             xycoords='figure pixels',
             fontsize=14)
plt.annotate(r"\textbf{\$2.32}",
             xy=(242, 5), 
             xycoords='figure pixels',
             fontsize=14)
plt.annotate(r"\textbf{\$1.00}",
             xy=(317, 5), 
             xycoords='figure pixels',
             fontsize=14)
plt.annotate(r"\textbf{\$0.42}",
             xy=(392, 5), 
             xycoords='figure pixels',
             fontsize=14)


# Save the figure
fig.savefig('twelve_model_results.png', bbox_inches='tight')
```
